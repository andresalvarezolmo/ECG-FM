```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the validation metrics from valid.csv
df = pd.read_csv("/home/aa2650/playground/ECG-FM/notebooks/outputs/2025-03-30/17-43-10/valid.csv")

# Create an "epoch" column if each row represents an epoch (starting from 1)
df["epoch"] = range(1, len(df) + 1)

# Define the target metrics to plot (update labels if necessary)
metrics = {
    "AUPRC": "auprc",
    "EM Accuracy": "em_accuracy",
    "F1 Score": "f1",
    "Accuracy": "accuracy",
    "AUROC": "auroc"
}

# Set font sizes
plt.rcParams.update({
    "font.size": 14,           # Base font size
    "axes.titlesize": 18,      # Title font size
    "axes.labelsize": 16,      # Axis label font size
    "xtick.labelsize": 12,     # X-tick label size
    "ytick.labelsize": 12,     # Y-tick label size
    "legend.fontsize": 12,     # Legend font size
    "figure.titlesize": 20     # Figure title font size (if used)
})

# Define colors using a professional colormap (tab10)
colors = plt.get_cmap("tab10")

plt.figure(figsize=(12, 7))
plt.style.use("seaborn-v0_8-whitegrid")

# Plot each metric with a unique color and markers for clarity
for i, (label, col) in enumerate(metrics.items()):
    plt.plot(df["epoch"], df[col], label=label, color=colors(i), linewidth=2)

plt.xlabel("Epoch")
plt.ylabel("Metric Value")
plt.title("ECG-FM Validation Performance Metrics Over Epochs")
plt.legend(loc="best")
plt.ylim(0.85, 1)
plt.xlim(0, 100)
plt.tight_layout()

# Define the output path
output_file = "/home/aa2650/playground/ECG-FM/results/validation_metrics.png"

# Save the figure in high quality (300 dpi)
plt.savefig(output_file, dpi=300, bbox_inches="tight")
plt.show()
```
